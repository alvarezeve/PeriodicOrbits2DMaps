## 游 Mapeo est치ndar

Este es un notebook que ejemplifica como usar el m칩dulo creado para calcular puntos peri칩dicos en mapeos de dos dimensiones usando el m칠todo de simetr칤as.

Primero debemos dar las ecuaciones del mapeo est치ndar en una manera especial
$$F_{x}= f(x,y)\\
F_{y}= g(x,y)$$
donde he omitido que las funciones pueden depender de par치metros. 

```julia
# definimos las funciones del mapeo est치ndar.
function EstandarX(x,y,풬,b)
    vx = x + y - 풬*sin(2*pi*x)/2*pi
    return vx
end

function EstandarY(x,y,풬,b)
    vy =  y - 풬*sin(2*pi*x )/2*pi 
    return vy
end
# notemos que las funciones tienen un par치metro b que est치 de m치s 
# como argumento pero que no se ocupa en las ecuaciones, esto es simple notaci칩n.
```

游댯 Definimos las simetr칤as conocidas del mapeo 
$$x=0\\
x=1/2\\
x=y/2\\
x=\frac{y+1}{2}$$

```julia
# definidas como funciones de y para que sean generales.
simetria1(y)= 0.0
simetria2(y)=1/2
simetria3(y)=y/2
simetria4(y)=(y+1)/2
```

游댯 Calculamos los puntos de periodo 2 en el intervalo $[0,1]$.
La notaci칩n es la siguiente:


`PeriodicosS(Funcion x del mapeo, Funcion y del mapeo, condicion de simetria, parametro1, parametro 2, periodo, intervalo)`

```julia
PuntosP2A = PeriodicosS(EstandarX,EstandarY,simetria1,풬,b,2,[-10.0,1.0])
PuntosP2B =PeriodicosS(EstandarX,EstandarY,simetria2,풬,b,2,[-20.0,10.0])
PuntosP2C = PeriodicosS(EstandarX,EstandarY,simetria3,풬,b,2,[-10.0,1.0])
PuntosP2D = PeriodicosS(EstandarX,EstandarY,simetria4,풬,b,2,[-10.0,1.0])
```

```julia
PuntosP3A = PeriodicosS(EstandarX,EstandarY,simetria1,풬,b,3,[-10.0,1.0])
PuntosP3B = PeriodicosS(EstandarX,EstandarY,simetria2,풬,b,3,[-10.0,10.0])
PuntosP3C = PeriodicosS(EstandarX,EstandarY,simetria3,풬,b,3,[-10.0,1.0])
PuntosP3D = PeriodicosS(EstandarX,EstandarY,simetria4,풬,b,3,[-10.0,1.0]);
```

游댮 Para filtrar los verdaderos puntos de periodo 2 lo que haremos es usar la siguiente funci칩n.

```julia
#definimos un mapeo con los par치metros fijos.
Mapeo(x,y) = [EstandarX(x,y,풬,0.1),EstandarY(x,y,풬,0.1)]

# definimos una funci칩n que itera el mapeo.
function MapeoIterador(p,n) #p punto inicial n= numero de iteraciones
    lista1 = [] 
    lista2 = []
    append!(lista1,mod(p[1],1)) # listas con las iteraciones
    append!(lista2,mod(p[2],1))
    for i in 1:n # comienza la iteraci칩n
        p = Mapeo(p[1],p[2]) 
        p = [mod(p[1],1),mod(p[2],1)] # como el mapeo no tiene definido implicito el mod 1 lo agregamos
        append!(lista1,p[1])
        append!(lista2,p[2])
    end
    return [lista1,lista2]
end
```

Unimos todos los puntos de periodo 2,3 encontrados en las diferentes simetrias. 

```julia
Puntos2px =vcat(PuntosP2A[1],PuntosP2B[1],PuntosP2C[1],PuntosP2D[1])
Puntos2py = vcat(PuntosP2A[2],PuntosP2B[2],PuntosP2C[2],PuntosP2D[2])
#Puntos2 = [Puntos2px, Puntos2py]
Puntos3 =[vcat(PuntosP3A[1],PuntosP3B[1],PuntosP3C[1],PuntosP3D[1]),
    vcat(PuntosP3A[2],PuntosP3B[2],PuntosP3C[2],PuntosP3D[2])];
```

游늷 Filtramos los puntos que son de periodo dos de los que no. 

```julia
function FilterPoints(Puntos,periodo)
        for i in 1:length(Puntos[1])
        iteraciones = MapeoIterador([Puntos[1][i],Puntos[2][i]],periodo+1) #iteramos los puntos encontrados
        if abs(mod(iteraciones[1][periodo+1],1) - mod(iteraciones[1][1],1)) <= 1e-90 && 
            abs(mod(iteraciones[2][periodo+1],1) - mod(iteraciones[2][1],1))  <= 1e-90 &&
            [iteraciones[1][1],iteraciones[2][1]]-[iteraciones[1][2],iteraciones[2][2]] != [0.0,0.0]
            # comparamos que despu칠s de ciertas iteraciones se vuelva al mismo punto y que no se vuelva antes
            print("El punto " ,i , " es de periodo ", periodo)
            print("\n")
       end
    end
end
```

```julia
function IterarMap(f, x_ini, n)   #Definimos una funci칩n para iterar el mapeo

    x = x_ini[1]                      #                                     #Damos condiciones iniciales
    y = x_ini[2]                          #

    iteradosMapX = [x]
    iteradosMapY = [y]               #Definimos dos listas que tendran los valores de cada par ordenado de theta y P, y agregamos las condiciones iniciales

    for i=0:n              #iniciamos un ciclo de iteraciones donde se calculan x_n, y_n y se agregan a lalista correspodiente
        F = f(x,y)

        push!(iteradosMapX,F[1])
        push!(iteradosMapY,F[2])

        x = F[1]
        y = F[2]

    end

    return iteradosMapX, iteradosMapY  #La funcion iterados regresa las listas que corresponden a la trayectoria del

end

function GraficarEstandarMap(f)
    n = 50
    #s=2*pi/14.
    s = 1/20.
    i=0.
    for p_i=0:s:1
        for x_i =0:s:1
            i=i+1
            x_ini=[p_i,x_i]
            a,b = IterarMap(f,x_ini,n)
            p = scatter(a,b,marker=".",s=0.1,color="gray")
            
        end
    end

end
```

```julia
#-------------------------------------------------------------------------------
"""
Estandar(x::Any, y::Any k::Real)

Funci칩n para el mapeo est치ndar con m칩dulo 2pi.
"""
function Estandar(x::Any,y::Any,k::Real)
    x_n =mod(x+y-k*sin(2*pi*x)/2pi,1)
    y_n =mod(y-k*sin(2*pi*x)/2pi,1)
    return [x_n,y_n]
end
```
